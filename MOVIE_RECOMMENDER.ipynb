```python
df_movies=df_movies[['movie_id','title','keywords','genres','overview','cast','crew']]
```

```python
import ast
```

"ast" is a module that is used to convert a string to a list with the help of which we can convert several columns of the df_movies to list type

```python
### we will create a function to convert string to a list

def convert(item):
  l=[]
  for i in ast.literal_eval(item):
    l.append(i['name'])
  return l
```

```python
### we will check for duplicate values

df_movies.duplicated().sum()
```

```python
### checking for missing values...

df_movies.isnull().sum()
```

*since there are only 3 missing values, we will drop those missing values*

```python
df_movies.dropna(inplace=True)
```

```python
df_movies['genres']=df_movies['genres'].apply(convert)
df_movies.head()
```

*thus we can see that the genre column is processed and looks clean*

```python
df_movies['keywords']=df_movies['keywords'].apply(convert)
```

```python
df_movies.iloc[0].keywords
```

```python
df_movies.head()
```

```python
df_movies['cast']=df_movies['cast'].apply(convert)
df_movies.head()
```

```python
df_movies['cast']=df_movies['cast'].apply(lambda x:x[0:3])
df_movies.head()
```

```python
df_movies['crew'][0]
```

```python
### we will convert the overview into a list

df_movies['overview']=df_movies['overview'].apply(lambda x:x.split())
```

```python
df_movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df_movies['crew'] = df_movies['crew'].apply(fetch_director)
```

```python
df_movies.head()
```

**thus we finally have the dataset with all the columns that we need in order to build the recommender system**

**now we have to replace all the spaces that are present in the list so that our model doesn't get confused between two same names**

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df_movies['cast'] = df_movies['cast'].apply(collapse)
df_movies['crew'] = df_movies['crew'].apply(collapse)
df_movies['genres'] = df_movies['genres'].apply(collapse)
df_movies['keywords'] = df_movies['keywords'].apply(collapse)
```

```python
df_movies.head()
```

```python
### we create tags by concatenating every column together

df_movies['tags']=df_movies['overview']+df_movies['keywords']+df_movies['genres']+df_movies['cast']+df_movies['crew']
```

```python
new_movies=df_movies.drop(columns=['keywords','genres','overview','cast','crew'])
```

```python
new_movies.head()
```

```python
### this is the final processed dataframe which we will be using

new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python
len(cv.get_feature_names())

### we can see the 5000 most frequent words that are present in the tags....
```

*now from the above list of 5000 words we can see that many of the words are same but are counted differently, which is a problem*

*So to overcome this we need to apply stemming to convert them to the same word...*

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity.shape

### similarity of one movie with 4806 movies
```

```python
similarity[0]

### similarity of like movies will be closer to 1 & unlike movies will be closer to 0
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
